```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
# Generate some sample data (contoh data)
data_women = df_women['Price'].values
# Plot histogram (membuat histogram)
sns.histplot(data_women, bins = 20,kde=True, color='blue')
plt.title('Histogram Data')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')
plt.show()
```

```python
from scipy.stats import mannwhitneyu

# Uji Mann-Whitney U
statistic, p_value = mannwhitneyu(data_men, data_women)

# Interpretasi hasil
alpha = 0.05
print("Nilai Uji Statistik:", statistic)
print("Nilai p-value:", p_value)

if p_value < alpha:
    print("Terdapat perbedaan yang signifikan antara harga baju pada kelompok laki-laki dan perempuan.")
else:
    print("Tidak ada perbedaan yang signifikan antara harga baju pada kelompok laki-laki dan perempuan.")
```
